## Visualization & Explanatory Analysis

### Pertanyaan 1: 

Berapakah jumlah rata-rata peminjam sepeda dalam setiap bulan ? - 10122099 - Muhammad Raffy Abdillah

```python
data_day['year_month'] = pd.to_datetime(data_day['dteday']).dt.to_period('M')

df_result = data_day.groupby('year_month').agg({
    'casual': 'mean',
    'registered': 'mean',
    'cnt': 'mean'
}).reset_index()

df_result['year_month'] = df_result['year_month'].astype(str)

plt.figure(figsize=(12, 6))
plt.plot(df_result['year_month'], df_result['casual'], label='Casual (avg)')
plt.plot(df_result['year_month'], df_result['registered'], label='Registered (avg)')
plt.plot(df_result['year_month'], df_result['cnt'], label='Cnt (avg)')
plt.xlabel('Tahun-Bulan')
plt.ylabel('Rata-Rata Peminjam')
plt.title('Rata-Rata Jumlah Peminjam Sepeda Per Bulan (2011-2012)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Pertanyaan 2: 

Pada season berapakah jumlah peminjaman sepeda terbanyak ? - 10122095 - Muhamad Snggih Prasetyo 

```python
total_cnt_by_season = data_day.groupby('season')['cnt'].sum().reset_index()
display(total_cnt_by_season)

x_coords = np.arange(len(list(total_cnt_by_season['season'])))

plt.bar(x_coords, total_cnt_by_season['cnt'], tick_label = total_cnt_by_season['season'],color=['#0000FF', '#0000FF', '#FF0000', '#0000FF'])
plt.xlabel('Musim')
plt.ylabel('Jumlah Peminjaman')
plt.title('Jumlah Peminjaman Berdasarkan Musim')

plt.show()
```

### Pertanyaan 3:

Apakah terdapat tren penggunaan sepeda yang jelas dari waktu ke waktu dalam dataset ini? - 10122088 - Azhar Fachrezi

```python
data_day['dteday'] = pd.to_datetime(data_day['dteday'])
by_date = data_day.groupby('dteday')['cnt'].sum()

plt.figure(figsize=(12, 6))
by_date.plot(kind='line', color='orange')
plt.title('Tren Penggunaan Sepeda dari Waktu ke Waktu')
plt.xlabel('Bulan-Tahun')
plt.ylabel('Total Penggunaan Sepeda (Cnt)')
plt.grid(True)
plt.show()
```

### Pertanyaan 4:

Berapakah rata-rata windspeed setiap bulan dalam tahun 2011-2012 ? - 10122093 - Muhammad Fahreza Thias Akhzan

```python
data_day['dteday'] = pd.to_datetime(data_day['dteday'])

data_day['year'] = data_day['dteday'].dt.year
data_day['month'] = data_day['dteday'].dt.month

data_day_filtered = data_day[(data_day['year'] >= 2011) & (data_day['year'] <= 2012)]

avg_windspeed_per_month = data_day_filtered.groupby(['year', 'month'])['windspeed'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(avg_windspeed_per_month['windspeed'], marker='o')
plt.title('Rata-rata Windspeed tiap Bulan (2011-2012)')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Windspeed')
plt.xticks(range(0, 24), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec',
                          'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

### Pertanyaan 5:

Berapa perbandingan jumlah peminjaman sepeda saat weekday ? - 10122116 - Haykal Fadhillah Suryana

```python
data_2011 = data_day[(data_day['yr'] == 0) & (data_day['weekday'].isin([0, 1, 2, 3, 4]))]
data_2012 = data_day[(data_day['yr'] == 1) & (data_day['weekday'].isin([0, 1, 2, 3, 4]))]

count_2011 = data_2011.groupby('weekday')['cnt'].sum()
count_2012 = data_2012.groupby('weekday')['cnt'].sum()

labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
plt.bar(labels, count_2011, width=0.4, label='2011', align='edge', color='blue')
plt.bar(labels, count_2012, width=-0.4, label='2012', align='edge', color='orange')

plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Peminjam')
plt.title('Perbandingan Jumlah Peminjam pada Weekday antara Tahun 2011 dan 2012')
plt.legend()
plt.show()
```

### Pertanyaan 6

Pengaruh weathersit terhadap rata-rata peminjaman sepeda ? - 10122094 - Mochamad Nabil Ramdhani

```python
rata_cnt_by_weathersit = data_day.groupby('weathersit')['cnt'].mean().reset_index()
display(rata_cnt_by_weathersit)

x_coords = np.arange(len(list(rata_cnt_by_weathersit['weathersit'])))

plt.bar(x_coords, rata_cnt_by_weathersit['cnt'], tick_label = rata_cnt_by_weathersit['weathersit'],color=['#FF0000', '#0000FF', '#0000FF'])
plt.title('Rata-Rata Peminjam Berdasarkan Weathersit')
plt.xlabel('Weathersit')
plt.ylabel('Rata-Rata Peminjaman Sepeda')
plt.show()
```

### Pertanyaan 7


Bagaimana perubahan kelembaban pada tanggal 06 Maret 2012 ? - 10122099 - Muhammad Raffy Abdillah

```python
filtered_data = data_hour[data_hour['dteday'] == '2012-03-06']
hrs = filtered_data['hr']
humidity_percent = filtered_data['hum'] * 100

plt.subplots(figsize=(10, 6))
plt.plot(hrs, humidity_percent, marker='o', color='blue')
plt.title('Kelembaban Terhadap Jam pada Tanggal 06-03-2012')
plt.xlabel('Jam')
plt.ylabel('Kelembaban (%)')
plt.xticks(range(24)) 
plt.grid(True)
    
plt.show()
```
